```python
import xarray as xr
from pathlib import Path
import copy
import pandas as pd
import json
import os
base_path = Path('/fs/site6/eccc/cmd/w/sbf000/reanalysis-era5-single-levels')
```

```python
file = 'reanalysis-era5-single-levels.nc'
ds = xr.open_dataset(str(base_path / file))
```

# Era5 Single levels

```python
file = 'reanalysis-era5-single-levels.nc'
ds = xr.open_dataset(str(base_path / file))
display(ds)
dicts = []
for k,v in ds.data_vars.items():
    mydict = copy.deepcopy(ds[k].attrs )
    mydict['name'] = k
    mydict['dtype'] = str(ds[k].data.dtype)
    mydict['status'] = 'ACTIVE'
    mydict['level'] = 13
    mydict['interpolation'] = 'near'
    
    dicts.append(mydict)

df = pd.DataFrame(dicts)
display(df)
df.rename(columns = {'long_name':'name', 'name':'name_alternate','dtype':'datatype'}, inplace = True)
df['description_short'] = df.name
df = df[['name', 'name_alternate', 'description_short', 'datatype', 'units', 'status', 'level', 'interpolation']]
df.loc[df.datatype == 'float32', 'datatype'] = 'fl'
display(df)
result = df.to_json(orient="records")
parsed = json.loads(result)
json_text = json.dumps(parsed, indent=4)
json_file = '../reanalysis-era5-single-levels.json'
if os.path.isfile(json_file):
    os.remove(json_file)
with open(json_file,'w') as fp:
    fp.write(json_text)
```
