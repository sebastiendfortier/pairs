```python
import fstpy.all as fstpy
import fstd2nc
import zarr
import xarray as xr
import pandas as pd
import os
from pathlib import Path
from shutil import rmtree
# pd.options.display.max_rows = 100
import rpnpy.librmn.all as rmn
rmn.fstopt(rmn.FSTOP_MSGLVL, rmn.FSTOPI_MSG_FATAL, setOget=0)

df = fstpy.StandardFileReader('/home/sbf000/ss5/source_data_5005_new.std', decode_metadata=True).to_pandas()

# df.date_of_validity.unique()

meta_df = fstpy.select_meta(df)

no_meta_df = df.loc[~df.nomvar.isin(meta_df.nomvar.unique())]


if os.path.exists('/home/sbf000/ss5/zarr_data/'):
    rmtree('/home/sbf000/ss5/zarr_data/')

groups = no_meta_df.groupby(['date_of_validity','date_of_observation','grid','vctype','nomvar'])

for (date_of_validity,date_of_observation,grid,vctype,nomvar), sub_df in groups:

    ndf = pd.concat([meta_df,sub_df],ignore_index=True)

    ndf = fstpy.metadata_cleanup(ndf)

    # print(date_of_validity.strftime('%Y%m%dT%H%M%SZ'),date_of_observation.strftime('%Y%m%dT%H%M%SZ'), grid, nomvar,vctype)

    # display(ndf)

    ds = fstd2nc.Buffer.from_fstpy(ndf).to_xarray()
    # display(ds)

    zarr_file = f"/home/sbf000/ss5/zarr_data/{date_of_validity.strftime('%Y%m%dT%H%M%SZ')}/{date_of_observation.strftime('%Y%m%dT%H%M%SZ')}/{grid}/{vctype}/{nomvar}"
    print(zarr_file)
    ds.to_zarr(zarr_file,compute=True)
```
