```python
from bs4 import BeautifulSoup
from pathlib import Path
import os
import re
import requests
import urllib.request

download_dir = Path('/home/sbf000/ss6/downloads')

DATASETS = {
'_DBLL_',
'_EATM_',
'_ETAL_',
'_HTGL_',
'_ISBL_',
'_MSL_',
'_NTAT_',
'_SFC_'
}

VARS = {
 '_4LFTX_',
 '_ABSV_',
 '_ACPCP_',
 '_ALBDO_',
 '_APCP06_',
 '_APCP12_',
 '_APCP18_',
 '_APCP24_',
 '_APCP30_',
 '_APCP36_',
 '_APCP42_',
 '_APCP48_',
 '_CAPE_',
 '_CWAT_',
 '_DEN_',
 '_DEPR_',
 '_DLWRF_',
 '_DPT_',
 '_DSWRF_',
 '_FPRATE_',
 '_GUST_',
 '_HGT_',
 '_HLCY_',
 '_HPBL_',
 '_ICEC_',
 '_IPRATE_',
 '_LAND_',
 '_LHTFL_',
 '_NLWRS_',
 '_NSWRS_',
 '_PRATE_',
 '_PRES_',
 '_PRMSL_',
 '_PTYPE_',
 '_RH_',
 '_RPRATE_',
 '_SDEN_',
 '_SDWE_',
 '_SFCWRO_',
 '_SHTFL_',
 '_SHWINX_',
 '_SKINT_',
 '_SNOD_',
 '_SOILM_',
 '_SOILVIC_',
 '_SPFH_',
 '_SPRATE_',
 '_TCDC_',
 '_THICK_',
 '_TMP_',
 '_TSOIL_',
 '_UGRD_',
 '_ULWRF_',
 '_USWRF_',
 '_VGRD_',
 '_VVEL_',
 '_WDIR_',
 '_WIND_',
 '_WTMP_'}

def get_files(run,passe):
    url = f'https://dd.weather.gc.ca/model_hrdps/continental/grib2/{run:02}/{passe:03}'
    ext = 'grib2'
    page = requests.get(url).text
    # print (page)
    soup = BeautifulSoup(page, 'html.parser')
    return [url + '/' + node.get('href') for node in soup.find_all('a') if node.get('href').endswith(ext)]

def filter_files():
    for r in [0,6,12,18]:
        if r == 0:
            passes = [0,24,48]
        elif r == 6:
            passes = [18,42]
        elif r == 12:    
            passes = [12,36]
        elif r == 18:    
            passes = [6,30]
        for p in passes:
            files = get_files(r,p)
            dset_filtered = []
            for dset in DATASETS:
                dset_filtered = dset_filtered + [f for f in files if dset in f] 

    var_filtered = []
    for var in VARS:
        var_filtered = var_filtered + [f for f in dset_filtered if var in f] 
    return var_filtered

def download_files(datamart_urls):
    if not download_dir.is_dir():
        os.makedirs(download_dir)

    files_to_process = []
    for url in datamart_urls:
        file = download_dir / re.sub('.*/(.*.grib2)',r'\1',url)
        files_to_process.append(file)
        if not file.is_file():
            urllib.request.urlretrieve(url, str(file))

datamart_urls = filter_files()

download_files(datamart_urls)

```
